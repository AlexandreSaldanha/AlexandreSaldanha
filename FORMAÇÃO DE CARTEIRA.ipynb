```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
portfolio_BR = yf.download(tickers= ('WEGE3.SA', 'EGIE3.SA', 'BBDC3.SA', 'LREN3.SA'), start= '2010-1-1') [['Adj Close']]
portfolio_BR
```

```python
portfolio_BR.plot(figsize = (13,7))
```

```python
portfolio_BR/portfolio_BR.iloc[0]*100
```

```python
(portfolio_BR/portfolio_BR.iloc[0]*100).plot(figsize = (14,6))
```

```python
rtn = (portfolio_BR/portfolio_BR.shift(1))-1
rtn
```

```python
weights = np.array([0.25,0.25,0.25,0.25])
np.dot(rtn,weights)
```

```python
an_rtns = rtn.mean()*252
an_rtns
```

```python
print(str(round(np.dot(an_rtns,weights),5)*100)+ '% a.a')
```

```python
weights_2 = np.array([0.4,0.1,0.1,0.4])
```

```python
print(str(round(np.dot(an_rtns,weights_2),5)*100)+ '% a.a')
```
